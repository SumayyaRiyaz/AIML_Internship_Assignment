```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [20000, 25000, 30000, 28000, 35000, 40000]
expenses = [12000, 15000, 18000, 16000, 20000, 22000]
profit = [s - e for s, e in zip(sales, expenses)]

plt.figure()
plt.bar(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()


plt.figure()
plt.pie(sales, labels=months, autopct="%1.1f%%")
plt.title("Sales Contribution by Month")
plt.show()


plt.figure()
plt.hist(sales, bins=5)
plt.title("Sales Distribution Histogram")
plt.xlabel("Sales Range")
plt.ylabel("Frequency")
plt.show()
```
