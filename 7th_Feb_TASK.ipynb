```python
#Task 1 : Password authenticator
password = input("Enter the password: ")

if len(password) < 8:
    print("Password must be at least 8 characters long")

elif "@" not in password and "_" not in password:

    print("Password must contain special characters(Access denied)")

elif password.isupper():
  print("password must contain lower case letters")

elif password.islower():
  print("password must contain upper case letters")

elif password == "Python_Exec@123":
    print("Access granted")

else:
    print("Access denied")
```
