```python
#2.Smart Input program
name = input("Enter Name:")
age = int(input("Enter Age:"))
hobby = input("Enter a Hobby:")
if(age < 12):
    print("You are a child full of energy. Great that you have " + hobby + " as your hobby. Keep growing!")
elif(13 <= age <= 19):
    print("You are a Teen with lots of curiosity. Enjoy " + hobby + " as your hobby & keep learning!")
elif(20 <= age <= 59):
    print("You are an Adult with responsibilities. Spend your time productively by adapting your hobby as"  + hobby)
else:
    print("You are a Senior citizen. Great that you spend time enjoying your favourite hobby as " + hobby)
```
