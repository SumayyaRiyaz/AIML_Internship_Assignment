```python
students={
    "Maryum":75,
    "Sarah":90,
    "Aliza":66,
    "Junaid":98,
    "Ayan":87
}
topper = max(students, key=students.get)
print("Topper:", topper)
print("Marks:", students[topper])
print("Average:", sum(students.values())/len(students))
for name, marks in students.items():
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    else:
        grade = "D"

    print(name, "-> Marks:", marks, "Grade:", grade)
```
